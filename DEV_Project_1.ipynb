```
plt.scatter(df['day'],df['balance'])
```

```
fig,axes = plt.subplots()
axes.stem(df['day'],df['balance'],use_line_collection = True, basefmt=' ')

plt.title('Bank')
plt.xlabel('day')
plt.ylabel('balance')
plt.xticks(df['day'])
plt.show()
```

```
sns.boxplot(x='day',y='balance',data=df)
plt.show()
```
